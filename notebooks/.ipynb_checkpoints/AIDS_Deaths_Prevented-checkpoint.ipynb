```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
art_df = pd.read_csv('../data/ART_data.csv')
```

```python
print(art_df.head())
```

```python
del art_df['Value Footnotes']
del art_df['Source']
del art_df['Unit']
```

```python
print(art_df.head())
```

```python
art_df.columns = ['Country', 'Subgroup', 'Year', 'No_of_Deaths']
print(art_df.head(20))
```

```python
afg_df = art_df[art_df['Country'] == 'Afghanistan']
print(afg_df)
plt.plot('Year', 'No_of_Deaths', data=afg_df[afg_df['Subgroup'] == 'upper estimate'], color = 'blue', linestyle = '--', label = 'upper est')
plt.plot('Year', 'No_of_Deaths', data=afg_df[afg_df['Subgroup'] == 'estimate'], color = 'blue', label = 'estimate')
plt.plot('Year', 'No_of_Deaths', data=afg_df[afg_df['Subgroup'] == 'lower estimate'], color = 'blue', linestyle = ':', label = 'lower est')
plt.title('Afghanistan AIDS deaths prevented with ART')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.xticks([2010, 2011, 2012, 2013, 2014])
plt.legend()
```

```python
def graph_country(country):
    """This function takes a country string as input and returns a graph of the number of AIDS deaths prevented with ART"""
    df = art_df[art_df['Country'] == country]
    plt.plot('Year', 'No_of_Deaths', data=df[df['Subgroup'] == 'upper estimate'], color = 'blue', linestyle = '--', label = 'upper est')
    plt.plot('Year', 'No_of_Deaths', data=df[df['Subgroup'] == 'estimate'], color = 'blue', label = 'estimate')
    plt.plot('Year', 'No_of_Deaths', data=df[df['Subgroup'] == 'lower estimate'], color = 'blue', linestyle = ':', label = 'lower est')
    plt.title(country +' AIDS deaths prevented with ART')
    plt.xlabel('Year')
    plt.ylabel('Number of Deaths')
    plt.xticks([2010, 2011, 2012, 2013, 2014])
    plt.legend()
    return plt.show()
```

```python
graph_country('Algeria')
```

```python
graph_country('Ethiopia')
```

```python
print(art_df['Country'].unique())
```

```python
graph_country('Bolivia (Plurinational State of)')
```

```python
for country in art_df['Country'].unique():
    graph_country(country)
```
